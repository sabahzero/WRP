```python
import gzip
import json

with gzip.open("biohackathon-prefilter.ndjson.gz", "r") as f:
   data = f.read()
   j = json.loads (data.decode('utf-8'))
```

```python
print(":ID,:LABEL,name")
for item in j:
    print(item["id"]+",Compound,"+item["labels"]["en"]["value"])
```

```python

import os

outfilename = 'biohackathon-prefilter.csv.gz'
with gzip.open(outfilename, 'wt') as f:
    f.write(":ID,:LABEL,name\n".encode("UTF-8"))
    for item in j:
        csvline = item["id"]+",Compound,"+item["labels"]["en"]["value"]+"\n"
        f.write(csvline.encode("UTF-8"))
```
